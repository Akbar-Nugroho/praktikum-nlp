# Soal 2

<br>Lakukan pembersihan meliputi penghapusan punctuation, angka dan karakter yang tidak penting menggunakan menggunakan regex.
<br>Simpan hasilnya menjadi file csv/excel.


```python
#Jawaban
```

```python
import pandas as pd
import re
```

```python
df = pd.read_csv('crawling-tweet-gibran-rakabuming.csv')
```

```python
# Kolom yang akan dibersihkan
kolom_teks = 'full_text'

# Fungsi untuk membersihkan teks dari punctuation, angka, dan karakter tidak penting
def bersihkan_teks(full_text):
    teks_bersih = re.sub(r'[^a-zA-Z\s]', '', str(full_text))
    return teks_bersih

# Terapkan fungsi pembersihan pada kolom
df[kolom_teks] = df[kolom_teks].apply(bersihkan_teks)
```

```python
df.to_csv('hasil_pembersihan_gibran_rakabuming.csv', index=False)
```

# Soal 3d

d.Lakukan stopword removing  berdasarkan hasil 3c, simpan hasilnya dalam bentuk csv/excel. (bobot:10)!

```python
#Jawaban
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
```

```python
df = pd.read_csv('slang_words_gibran_rakabuming.csv')
```

```python
# Download stopwords jika belum terinstal
nltk.download('stopwords')

# Ambil daftar stopwords bahasa Indonesia
stopwords_indonesia = set(stopwords.words('indonesian'))

# Fungsi untuk menghapus stopwords
def hapus_stopwords(tokens):
    tokens_tanpa_stopwords = [kata for kata in tokens if kata.lower() not in stopwords_indonesia]
    return tokens_tanpa_stopwords

# Hapus stopwords dari kata-kata slang
df['Tokens_Tanpa_Stopwords'] = df['slang_words'].apply(hapus_stopwords)
```

```python
df
```

```python
df.to_csv('stopword_removing_gibran_rakabuming.csv')
```

```python
import pandas as pd
import nltk
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
df = pd.read_csv('stopword_removing_gibran_rakabuming.csv')
```

```python
# Inisialisasi stemmer Sastrawi untuk Bahasa Indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
# Fungsi untuk melakukan stemming pada tokens
def stemming(tokens):
    stemmed_tokens = [stemmer.stem(kata) for kata in tokens]
    return stemmed_tokens

# Melakukan stemming pada kata-kata slang
df['Stem'] = df['Tokens_Tanpa_Stopwords'].apply(stemming)
```
